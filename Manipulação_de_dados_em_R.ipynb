```r
#Grafico de pizza de porcentagens
valores <- c(1,2,3,4)
lbls <- c("AC", "AL", "AM", "AP")
percent <- round( valores/sum(valores)*100)

rotulos <- paste(rotulos, percent)
torulos <- paste(rotulos,"%",sep="")

pie(valores,labels = rotulos, col=rainbow(length(rotulos)), main="Grafico de porcentagens")

```
